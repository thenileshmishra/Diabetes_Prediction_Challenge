# Basic imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv("/Users/nileshmishra/Kaggel/data/raw/train.csv")  
df.head()
```

## Basic Info

```python
print("Shape:", df.shape)
df.info()
df.describe().T
```

## Check Missing & Duplicates

```python
df.isnull().sum()

print("Duplicate rows:", df.duplicated().sum())
```

## Target Variable Overview

```python
df['diagnosed_diabetes'].value_counts(normalize=True)*100

sns.countplot(data=df, x='diagnosed_diabetes')
plt.title("Diabetes Class Distribution")
plt.show()
```

## Separate Numerical & Categorical Columns

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Numerical:", num_cols)
print("Categorical:", cat_cols)
```

# Univariate Analysis

## Numerical Feature Distributions

```python
plt.figure(figsize=(16,12))
for i,col in enumerate(num_cols,1):
    plt.subplot(5,6,i)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

## Categorical Feature Distributions

```python
plt.figure(figsize=(16,10))
for i,col in enumerate(cat_cols,1):
    plt.subplot(3,3,i)
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=45)
    plt.title(col)
plt.tight_layout()
plt.show()
```

# Bivariate Analysis vs Diabetes

## Numerical vs Target

```python
plt.figure(figsize=(16,12))
for i,col in enumerate(num_cols,1):
    plt.subplot(5,6,i)
    sns.boxplot(x='diagnosed_diabetes', y=col, data=df)
    plt.title(col + " vs Diabetes")
plt.tight_layout()
plt.show()
```

## Categorical vs Target

```python
plt.figure(figsize=(16,10))
for i,col in enumerate(cat_cols,1):
    plt.subplot(3,3,i)
    prop = (df.groupby(col)['diagnosed_diabetes']
              .mean()
              .reset_index())
    
    sns.barplot(data=prop, x=col, y='diagnosed_diabetes')
    plt.xticks(rotation=45)
    plt.title(col + " vs Diabetes Rate")
plt.tight_layout()
plt.show()
```

## Correlation & Multicollinearity

```python
plt.figure(figsize=(14,10))
corr = df[num_cols].corr()
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_target = corr['diagnosed_diabetes'].sort_values(ascending=False)
corr_target
```

## Outlier Detection

```python
plt.figure(figsize=(16,12))
for i,col in enumerate(num_cols,1):
    plt.subplot(5,6,i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

## Using IQR

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5*IQR)) | (df[col] > (Q3 + 1.5*IQR))]
    print(col, "Outliers:", len(outliers))
```
